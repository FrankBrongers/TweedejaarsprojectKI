```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

# Averaging

```python
img = cv2.imread("testField.png")
blur = cv2.blur(img,(100,100))
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
# cv2.imshow('image',img)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

# Gaussian, pretty oke, hopelijk goed genoeg voor edge detection

```python
blur =cv2.GaussianBlur(img,(155,155),0)
#  cv2.blur(img,(100,100))
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

# Bilateral Fcking traag

```python
blur = cv2.bilateralFilter(img,50,750,750)
# plt.subplot(121),plt.imshow(img),plt.title('Original')
# plt.xticks([]), plt.yticks([])
# plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
# plt.xticks([]), plt.yticks([])
# plt.show()
plt.imshow(blur)
```
